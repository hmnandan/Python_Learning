```python


lst = [2,2,1,1,1,2,2]
hm={}

for i in lst:
  if i in hm:
    hm[i]+=1
  else:
    hm[i]=1

print(hm)

from collections import Counter

c = Counter(lst)
print(c)

for i,j in c.items():
  print(i,j)

for i,j in c.items():
  if j==max(c.values()):
    print(i)
```
